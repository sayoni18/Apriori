```python
import pandas as pd
import numpy as np
```

```python
with open("F:/datascience/Python/Market_Basket_Optimisation.csv") as F:
    D=[x.strip() for x in F]
```

```python
A=[]
for i in D:
    A.append(i.split(','))
```

```python
A[0:5]
```

```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
TE=TransactionEncoder()
new=TE.fit(A).transform(A)
new
new.astype('int')
```

```python
B=pd.DataFrame(new,columns=TE.columns_)
```

```python
B.head()
```

```python
frequent_items=apriori(B,min_support=0.02,use_colnames=True)
```

```python
frequent_items.sample(10,random_state=40)
```

```python
rules=association_rules(frequent_items,metric="lift",min_threshold=1)
```

```python
rules.sample(5)
```

```python
rules.sort_values('confidence',ascending=False)
```
